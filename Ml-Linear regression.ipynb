Predict canada's per capita income in year 2020. We will be using  canada_per_capita_income.csv file. Using this we will build a regression model and predict the per capita income fo canadian citizens in year 2020


Problem Statement: Given above data building a machine learning model that can predict the per capita income fo canadian citizens in year 2020


```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/canada_per_capita_income.csv')
df.head(10)
```

```python
# @title year vs per capita income (US$)

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='year', y='per capita income (US$)', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
new_df = df.drop('per capita income (US$)',axis='columns')
new_df.head(10)
```

```python
canada_per_capita_income = df[('per capita income (US$)')]
canada_per_capita_income.head(10)
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,canada_per_capita_income)
```

```python
per_capita_income_in_2020 = reg.predict([[2020]])
print(per_capita_income_in_2020)
```

**Using this linear regression model we were able to predict the per capita income of canadian citizen in the year 2020 is 41288.69409442**
